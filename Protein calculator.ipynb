Balanced Protein calculator


```python
import pandas as pd

namel=[]
proteinl=[]
tproteinl = []
pricel=[]
pgl = []
p3gl = []
cl = []
p100cl = []
data = {
    'Name':namel,
    'Protein': proteinl,
    'Total Protein': tproteinl,
    'Price': pricel,
    'Price per g':pgl,
    'Price per 30g': p3gl,
    'Total Calories': cl,
    'Price per 100 calories': p100cl,
}

headers = list(data.keys())


def ins (n,pro,tpro,pri,pg,p3g,tc,p100c):
    namel.append(n)
    proteinl.append(pro)
    tproteinl.append(tpro)
    pricel.append(pri)
    pgl.append(pg)
    p3gl.append(p3g)
    cl.append(tc)
    p100cl.append(p100c)
    return
def price_per_gram (protein, price):
    return price/protein
def price_per_30g (ppg):
    return ppg*30
def price_per_100_calories (calories,price):
    return (price/calories)*100
    
pd.set_option('display.max_rows',500)

pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)
    
    
```

```python
while name!=0: 
    name = input('name') 
    if name == 0:
        break
    protein = int(input('how much protein'))
    calories = int(input('calories'))
    servings = float (input('servings'))
    price = float(input('price'))

    
    total_protein = protein*servings
    total_calories = calories*servings
    
    pg = price_per_gram(total_protein,price)
    p3g = price_per_30g(pg)
    p100c = price_per_100_calories(total_calories,price)
    
    ins(name,protein,total_protein,price,pg,p3g,total_calories,p100c)
    df = pd.DataFrame(data,columns = headers)
    print (df)
```
